## Kontur Algılama
Dış ve İç Konturlar

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
img = cv2.imread('/content/geometry.png')
img_gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
ret, thresh = cv2.threshold(img_gray, 20, 255, 0)
plt.imshow(thresh,cmap='gray')
```

**Konturları Bul**

Bir görüntüdeki konturları geri döndürür ve çağrılan RETR yöntemine bağlı olarak harici, dahili veya her ikisini de geri alabilir:

* cv2.RETR_EXTERNAL:Yalnızca dış konturları çıkarır
* cv2.RETR_CCOMP: İki seviyeli bir hiyerarşide düzenlenmiş hem iç hem de dış konturları çıkarır
* cv2.RETR_TREE: Bir ağaç grafiğinde düzenlenmiş hem iç hem de dış konturları çıkarır
* cv2.RETR_LIST: Herhangi bir iç/dış ilişki olmaksızın tüm konturları çıkarır

```python
contours, hierarchy = cv2.findContours(thresh, cv2.RETR_CCOMP, cv2.CHAIN_APPROX_SIMPLE)
```

```python
hierarchy
```

```python
i = img.copy()
cv2.drawContours(i, contours, contourIdx=-1, color=(0,255,0), thickness=3)
plt.imshow(i)
```

```python
i = img.copy()
cv2.drawContours(i, contours, 3, (0,255,0), 3)
plt.imshow(i)
```

```python
i = img.copy()
cnt = contours[4]
cv2.drawContours(i, [cnt], 0, (0,255,0), 3)
plt.imshow(i)
```
