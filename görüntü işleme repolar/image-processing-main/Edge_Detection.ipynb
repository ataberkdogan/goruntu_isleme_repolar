## Canny Kenar Tespiti
- gürültüyü gidermek amacıyla görüntüyü yumuşatmak için Gauss filtresi uygulayın
- görüntünün yoğunluk gradyanını bulun
- Kenar algılamaya yönelik yapay yanıttan kurtulmak için maksimum olmayan bastırma uygulayın
- potansiyel kenarları belirlemek için çift eşik uygulayın
- histerezis yoluyla izleme kenarı: Zayıf olan ve güçlü kenarlara bağlı olmayan tüm diğer kenarları bastırarak kenar tespitini sonlandırın

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
img = cv2.imread('/content/peppers_color.tif')
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
plt.imshow(img)

```

```python
edges = cv2.Canny(image=img, threshold1=127, threshold2=127)
plt.imshow(edges)
```

```python
edges = cv2.Canny(image=img, threshold1=0, threshold2=255)
plt.imshow(edges)
```

```python
med_val = np.median(img)
# Alt sınır, medyan değerin %0 veya %70'idir (hangisi daha yüksekse)
lower = int(max(0, 0.7* med_val))
# Üst sınır medyan değerinin %255 veya %30 üzerindedir (hangisi daha düşükse)
upper = int(min(255,1.3 * med_val))
```

```python
edges = cv2.Canny(image=img, threshold1=lower , threshold2=upper)
plt.imshow(edges)
```

```python
blurred_img = cv2.blur(img,ksize=(5,5))
edges = cv2.Canny(image=blurred_img, threshold1=lower , threshold2=upper)
plt.imshow(edges)
```
