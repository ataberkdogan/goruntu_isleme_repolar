## Görüntü Piramidi

1. Gauss Piramidi
- Görüntüleri alt örneklemek için kullanılır
- Alt düzey (daha yüksek çözünürlüklü) görüntüdeki ardışık satır ve sütunların kaldırılmasıyla oluşturulur
- üst seviyedeki her piksel, alttaki seviyedeki 5 pikselin Gauss ağırlıklarıyla katkısıyla oluşur
- alan orijinal alanın dörtte birine düşer *Oktav
2. Laplace Piramidi
- Piramidin alt kısmındaki (daha az çözünürlüklü) bir görüntüden üst örneklenmiş bir görüntüyü yeniden oluşturmak için kullanılır
- Gauss Piramitlerinden oluşmuştur
- yalnızca kenar görüntüleri, öğelerinin çoğu sıfırdır.

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
from google.colab.patches import cv2_imshow
```

```python
img = cv2.imread("/content/Lenna.png")
```

`pyrUp()` function increases the size to double of its original size
`pyrDown()` function decreases the size to half

```python
layer = img.copy()

for i in range(4):
    plt.subplot(2, 2, i + 1)
    layer = cv2.pyrDown(layer)
    imgCvt = cv2.cvtColor(layer, cv2.COLOR_BGR2RGB)
    # BGR'de temsil edilen OCV görüntüsünün dönüştürülmesi
    plt.imshow(imgCvt)
    cv2_imshow(layer)  #"str(i)"
    cv2.waitKey(0)


cv2.destroyAllWindows()
```
